```python
#medidas de posição e disperção
#dist. normal
import numpy as np
#produz 500 numeros aleatórios de média 0 e desvio padrão 1
np.random.seed(20210000)
x=np.random.normal(0,1,500) #dist. uniforme
media=np.mean(x)
d_pad=np.std(x,ddof=1)
print(f'média={media:.2f} e desvio padrão={d_pad:.2f}')
```

```python
# super histograma 
#dist. normal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#produz N números aleatórios de média 3 e desvio padrão 2
N=10000;
np.random.seed(20210000)
x=np.random.normal(3,2,N) #dist. uniforme
media=np.mean(x)
d_pad=np.std(x,ddof=1)

plt.rcParams.update({'font.size': 12})
n, bins,p  =plt.hist(x,bins=40,edgecolor='black', linewidth=1.2)
plt.xlabel('valores')
plt.ylabel('frequência')
plt.text(-4.7, 700, f'$\mu={media:.2f},\ \sigma={d_pad:.2f}$')
plt.axvline(media, color='r', linestyle='dashed', linewidth=2)
plt.axvline(media+d_pad, color='g', linestyle='dashed', linewidth=1)
plt.axvline(media-d_pad, color='g', linestyle='dashed', linewidth=1)
gauss = N*np.median(np.diff(bins))*norm.pdf(bins, media, d_pad)
plt.plot(bins, gauss, 'k-', linewidth=2)
```

```python
#calculando probabilidades
# calcula a probabilidade de encontrar um número entre A e B
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
A=0;
B=1;
P_AB = norm.cdf(B)-norm.cdf(A)
print(f'prob entre {A:.1f} e {B:.1f} = {P_AB:.3f}')
bins=np.linspace(-5,5,1000)
gauss = norm.pdf(bins, 0, 1)
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.plot(bins, gauss, 'b--', linewidth=2)
plt.xlabel('valores')
plt.ylabel('densidade de probabilidade (pdf)')
px=bins[(bins>A)&(bins<B)]
plt.fill_between(px,gauss[(bins>A)&(bins<B)],color='k',alpha=0.2)
```

```python
#calculando probabilidades
# calcula a probabilidade de encontrar um número entre A e B
# dist uniform
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
A=0.2;
B=0.4;
P_AB = uniform.cdf(B)-uniform.cdf(A)
print(f'prob entre {A:.1f} e {B:.1f} = {P_AB:.3f}')
bins=np.linspace(-0.3,1.3,1000)
unif = uniform.pdf(bins, 0, 1)
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.plot(bins, unif, 'b--', linewidth=2)
plt.xlabel('valores')
plt.ylabel('densidade de probabilidade (pdf)')
px=bins[(bins>A)&(bins<B)]
plt.fill_between(px,unif[(bins>A)&(bins<B)],color='k',alpha=0.2)
```

```python
#exercício 2: Cálculo de probabilidade
#distribuição exponencial
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
A=1;
B=2;
P_AB = expon.cdf(B)-expon.cdf(A)
print(f'prob entre {A:.1f} e {B:.1f} = {P_AB:.3f}')
bins=np.linspace(-1,7,1000)
expn = expon.pdf(bins)
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.plot(bins, expn, 'b--', linewidth=2)
plt.xlabel('valores')
plt.ylabel('densidade de probabilidade (pdf)')
px=bins[(bins>A)&(bins<B)]
plt.fill_between(px,expn[(bins>A)&(bins<B)],color='k',alpha=0.2)
```
